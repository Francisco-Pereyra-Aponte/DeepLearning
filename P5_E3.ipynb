```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#open 0-condor.png
img = plt.imread("0-condor.png")
plt.imshow(img)

b=0.0001
m=0.2/200

im = []

for i in range(201):
    beta = b + m*i
    e = np.random.normal(0, 1, img.shape)
    im.append(img * np.sqrt(1 - beta) + e * np.sqrt(beta))

index = [0,3,10,30,100,200]
fig, axes = plt.subplots(2, 3, figsize=(7,5))
for i, ax in zip(index, axes.flatten()):
    ax.imshow(im[i])
    ax.axis("off")
    ax.set_title(f"t={i}")
```
